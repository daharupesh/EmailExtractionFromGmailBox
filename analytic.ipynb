```python
original = series.drop_duplicates()
```

```python
df = pd.DataFrame(original)
```

```python
df["Name"]= df['Sender'].str.split('@').str[0]
```

```python
df.reset_index(drop=True)
```

```python
df.reset_index(drop=True)
```

```python
Order = ["Name" ,"Sender"]
```

```python
new_df = df[Order]
```

```python
new_df
```

```python
new_df.rename(columns={'Sender': 'Email_ID'}, inplace=True)
```

```python
new_df
```

```python
new_df['Name'] = new_df['Name'] \
    .str.replace('[._,-]', ' ', regex=True) \
    .str.title()
```

```python

df_sorted = new_df.sort_values(by='Name')
```

```python
df_sorted.reset_index(drop=True,inplace=True)
```
